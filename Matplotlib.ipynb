```python
sample_df=batter.head(100).sample(25,random_state=10)
sample_df
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
```

```python
plt.figure(figsize=(20,15))
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
plt.figure(figsize=(20,15))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

plt.axhline(130,color='red')
plt.axvline(30)
for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
fig,ax=plt.subplots(figsize=(15,9))
ax.scatter(batter['avg'],batter['strike_rate'],color='red')
ax.set_title('Average Runs Vs Strike Rate')
ax.set_xlabel('Average Runs')
ax.set_ylabel('Strike Rate')
fig.show()
```

```python
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(15,9))

ax[0].scatter(batter['avg'],batter['strike_rate'],color='red')
ax[1].scatter(batter['runs'],batter['strike_rate'],color='yellow')

ax[0].set_xlabel('Average Runs')
ax[0].set_ylabel('Strike Rate')
ax[0].set_title('Average Runs Vs Strike Rate')
ax[1].set_title('Runs Vs Strike Rate')
ax[1].set_xlabel('Runs')
ax[1].set_ylabel('Strike Rate')
fig.show()
```

```python
fig=plt.figure()
ax1=fig.add_subplot(2,1,1)
ax1.scatter(batter['avg'],batter['strike_rate'],color='red')
ax2=fig.add_subplot(2,1,2)
ax2.scatter(batter['runs'],batter['strike_rate'],color='yellow')
```

```python
from matplotlib import projections
fig=plt.figure(figsize=(15,9))

ax=plt.subplot(projection='3d')
ax.scatter3D(batter['runs'],batter['avg'],batter['strike_rate'])
ax.set_title('IPL Batsman Analysis')
ax.set_xlabel('Runs')
ax.set_ylabel('Avg Runs')
ax.set_zlabel('Strike Rate')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
xx,yy=np.meshgrid(x,y)
xx,yy
```

```python
xx.shape
```

```python
z=xx**2+yy**2
z.shape
```

```python
fig=plt.figure(figsize=(12,7))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z=np.sin(xx)+np.sin(yy)
fig=plt.figure(figsize=(12,7))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig=plt.figure(figsize=(12,7))
ax=plt.subplot()
p=ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig=plt.figure(figsize=(12,7))
ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
