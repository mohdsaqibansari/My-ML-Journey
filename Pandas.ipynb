```python
# Find number of movies starting with A in each genre
# mask=imdb['Series_Title'].str[0]=='A'
# x=imdb[mask]
# x.groupby('Genre')['Series_Title'].size()

def num_of_mov(x):
  mask=imdb['Series_Title'].str[0]=='A'
  x=imdb[mask]
  return x['Series_Title'].str.len()

genre.apply(num_of_mov)
```

```python
def movie_rank(group):
  group['Movie_Rank']=group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
genre.apply(movie_rank)
```

```python
# Find Normalized IMDB Rating group wise
def normal_rating(group):
  group['IMDB_Normalized']=(group['IMDB_Rating']-group['IMDB_Rating'].min())/(group['IMDB_Rating'].max()-group['IMDB_Rating'].min())
  return group
```
